TERNAS PITAGORICAS

```python
x=[i**2 for i in range(1,1001)]
y=[i**2 for i in range(1,1001)]
z=[i**2 for i in range(1,1001)]

def terna(a,b,c):
    ternas_Pitagoricas=[]
    if a+b==c:
        ternas_Pitagoricas.append((a,b,c))
        print(a,b,c)
for i in range(1000):
    for j in range(1000):
        for k in range(1000):
            terna(x[i],y[j],z[k])
```
